```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
url = "https://en.wikipedia.org/wiki/Chupacabra"
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup
```

```python
hyperlinks = soup.find_all('a')
num_hyperlinks = len(hyperlinks)
```

```python
print(f"Number of hyperlinks in the article: {num_hyperlinks}")
```

```python
def count_word_occurrences(url, word):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        text_content = soup.get_text()
        occurrences = len(re.findall(r'\b{}\b'.format(re.escape(word)), text_content, flags=re.IGNORECASE))
        return occurrences
    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
        return None
wiki_url = "https://en.wikipedia.org/wiki/Chupacabra"
word_to_find = 'or'
occurrences = count_word_occurrences(wiki_url, word_to_find)
if occurrences is not None:
    print(f"The word '{word_to_find}' occurs {occurrences} times in the article.")
```
